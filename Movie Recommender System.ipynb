```python
# genres ##id
# keywords
# title ##overview
# production_companies ##release_date
# cast #crew


movies = movies[['id','genres','keywords','title','overview','cast','crew','release_date','production_companies']]
```

```python
import ast 
def dp(movie):
    L=[]
    for i in ast.literal_eval(movie):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(dp)
```

```python
movies.head(1)
```

```python
movies[['keywords']].values
```

```python
movies['keywords'] = movies['keywords'].apply(dp)
```

```python
movies['production_companies'] = movies['production_companies'].apply(dp)
```

```python
movies.head()
```

```python
movies['cast'][0]
```

```python
def convert(obj):
    N = []
    count = 0
    for i in ast.literal_eval(obj):
        if count != 2:
            N.append(i['name'])
            count +=1
        else:
            break
    return N
```

```python
movies['cast'] = movies['cast'].apply(convert)
```

```python
movies.head(1)
```

```python
def fetch(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch)
```

```python
 movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(1)
```

```python
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['production_companies'] = movies['production_companies'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']=  movies['overview'] + movies['cast'] + movies['crew'] + movies ['genres'] + movies['keywords'] + movies['production_companies'] 
```

```python
movies.head(1)
```

```python
new_df = movies[['id','title', 'tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(obj):
    V=[]
    for i in obj.split():
        V.append(ps.stem(i))
    return " ".join(V)
    
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

# Finding Similarities
using Vectorisation


```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features= 5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
new_df.head(1)
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similar = cosine_similarity(vectors)
```

```python
similar
```

```python
list(enumerate(similar[0]))
```

```python
sorted(list(enumerate(similar[0])), reverse=True, key= lambda x:x[1])[1:9]
```

```python
def recommend(movie):
    index= new_df[new_df['title']== movie].index[0]
    distance=similar[index]
    m_list= sorted(list(enumerate(distance)), reverse=True, key= lambda x:x[1])[1:6]

    for i in m_list:
        #print(i[0])
        print(new_df.iloc[i[0]].title)
        
```

```python
recommend('The Dark Knight')
```
